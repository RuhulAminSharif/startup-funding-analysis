### **Finding the inappropriate dates**

```python
import re

def error_dates(df):
    dates = df.date.unique()
    # print(dates)
    # len(dates)
    wrong_dates = []
    for date in dates:
        if not re.match('^[0-9]{2}/[0-9]{2}/[0-9]{4}$', date):
            wrong_dates.append(date)
    return wrong_dates
print(error_dates(df))
```

```python
def clean_startup_verticals(df, column='vertical', manual_map=None, encoding_map=None, acronyms=None):
    """
    Cleans vertical column in-place using user-provided parameters.
    """
    if encoding_map:
        for pattern, replacement in encoding_map.items():
            df[column] = df[column].str.replace(pattern, replacement, regex=True)
    
    df[column] = df[column].str.replace(r'\\+[Nn]', ' ', regex=True)
    df[column] = df[column].str.replace(r'\s+', ' ', regex=True).str.strip()

    if manual_map:
        df[column] = df[column].replace(manual_map)

    if acronyms:
        for pattern, replacement in acronyms.items():
            df[column] = df[column].str.replace(fr'\b{pattern}\b', replacement, regex=True, flags=re.IGNORECASE)

    return df
```
